```python
# pandas를 통해 parquet 파일을 읽고 데이터를 Dataframe 형태로 로드
import pandas as pd


df = pd.read_parquet('./res/rag-eval-mini/data/korean_v1-00000-of-00001.parquet')
df
```

```python
print(df.iloc[0]['contexts'])
```

```python
# re(regular expression): 정규표현식
import re 

def extract_contexts(system_prompt):
    # context내에 조건을 만족하는 텍스트를 찾아 context_match에 저장
    context_match = re.search(r"\['(.*)'\]", system_prompt, re.DOTALL)
    if not context_match:
        return []

    # 정규표현식에서 첫번째 괄호에 묶인 그룹
    context_text = context_match.group(1)

    contexts = [context_text]
    
    # 불필요한 메타데이터 제거
    cleaned_contexts = []
    for context in contexts:
        lines = context.split('\n')
        cleaned_lines = []
        
        for line in lines:
            if '☞' in line:
                line = line.split('☞')[0].strip()
            cleaned_lines.append(line)
        
        cleaned_context = '\n'.join(cleaned_lines).strip()
        cleaned_contexts.append(cleaned_context)

    return cleaned_contexts

system_prompt = df.iloc[0]['contexts']
print(system_prompt)

extracted_contexts = extract_contexts(system_prompt)

for i, context in enumerate(extracted_contexts, 1):
    print(f"Context {i}:")
    print(context)
    print("-" * 50)
```

```python
# system 데이터 전처리
contexts = []
for i in range(len(df)):
    system_prompt = df.iloc[i]['contexts']
    extracted_contexts = extract_contexts(system_prompt)
    contexts.append(extracted_contexts)
```

```python
# 질문과 답변 데이터 전처리 
questions = [df.iloc[i]['question'] for i in range(len(df))]
answers = [df.iloc[i]['ground_truth'] for i in range(len(df))]
```

```python
#pickle 모듈을 활용해 데이터를 파일로 직렬화
import pickle

rag_data = {
    'questions': questions,
    'contexts': contexts,
    'answers': answers
}

with open('./res/rag_data.pkl', 'wb') as f:
    pickle.dump(rag_data, f)
```
